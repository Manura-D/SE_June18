c) Extract best-selling music artists from https://en.wikipedia.org/wiki/List_of_best-selling_music_artists

```python
import pandas as pd
import html5lib 
import numpy as np
import matplotlib.pyplot as plt

BestSelArt = pd.read_html('https://en.wikipedia.org/wiki/List_of_best-selling_music_artists', header=0, parse_dates=False)
BestSelArt

#BestSelArt[0].head(2)
#BestSelArt[1].head(2)
#BestSelArt[2].head(2)
#BestSelArt[3].head(2)
#BestSelArt[4].head(2)
#BestSelArt[5].head(2)

#BestSelArt[6].head(2)
#BestSelArt[7].head(2)
#BestSelArt[8].head(2)
#BestSelArt[9].head(2)

#BestSelArt[0].append(BestSelArt[1], ignore_index=True)

# Best Selling Artist List
ArtistList = [ BestSelArt[i]["Artist"][j] for i in range(0,6) for j in range(0,len(BestSelArt[i])) ]

# Best Selling Artist List with period
ArtistList2 = [ [BestSelArt[i]["Artist"][j] , BestSelArt[i]["Period active"][j]]  for i in range(0,6) for j in range(0,len(BestSelArt[i])) ]

print(ArtistList)
print("\n")
col = ("Artist", "ActPeri")
ArtistList2 = pd.DataFrame(ArtistList2, columns= col )
display(ArtistList2)
```

d) Find out how many best-selling music artists are from 1990s

```python
import re
Act = [ re.sub('present','2018',re.sub(',','',re.sub('\[.*\]','',x))) for x in ArtistList2["ActPeri"] ]

Fflg = [False] * len(Act)
for i in range(0,len(Act)):
    for spl in Act[i].split(' '):
        y = spl.split('–')
        if len(y)==2:
            flg = (int(y[0]) <= 1990) & (int(y[1]) >= 1990)
        else :
            flg = y[0][0:3] == '199'
        #print(y)
        #print(flg)
        Fflg[i] = Fflg[i] or flg
        #print("Fflg" , Fflg[i])

print("best-selling music artists are from 1990s")
display(ArtistList2[Fflg])
```

```python
# Using data feame manipulation
col = ("Artist", "ActPeri")
ArtistList2 = pd.DataFrame(ArtistList2, columns= col )
#ArtistList2.ActPeri.str[-4:]
ArtistList2['ActPeri'] = ArtistList2['ActPeri'].str.replace('present','2018')
ArtistList2['ActPeri'] = ArtistList2['ActPeri'].str.replace(',','')

ArtistList2[['Year1','Year2']] = ArtistList2['ActPeri'].str.split('[',expand=True )[0].str.split(' ',expand=True)[0].str.split('–',expand=True)
ArtistList2[['Year3','Year4']] = ArtistList2['ActPeri'].str.split('[',expand=True )[0].str.split(' ',expand=True)[1].str.split('–',expand=True)

ArtistList2['Year3'][ArtistList2['Year3'].isnull()] = ArtistList2['Year1'][ArtistList2['Year3'].isnull()]
ArtistList2['Year4'][ArtistList2['Year4'].isnull()] = ArtistList2['Year3'][ArtistList2['Year4'].isnull()]

for i in range (0,len(ArtistList2)):
    if (int(ArtistList2['Year1'][i]) <= 1990) & (int(ArtistList2['Year2'][i]) >= 1990) \
    | (int(ArtistList2['Year3'][i]) <= 1990) & (int(ArtistList2['Year4'][i]) >= 1990):
        print (ArtistList2['Artist'][i],ArtistList2['ActPeri'][i])
```
